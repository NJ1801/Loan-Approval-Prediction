```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostClassifier  # Import AdaBoostClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import OneHotEncoder
```

```python
data = pd.read_csv(r"C:\Users\USER\Downloads\loan_sanction.csv")
data
```

```python
# Select features and target variable
selected_features = [
   'Loan_ID','Gender','Married','Dependents','Education','Self_Employed','ApplicantIncome','CoapplicantIncome',
    'LoanAmount','Loan_Amount_Term','Credit_History','Property_Area'
]
```

```python
# Extract features and target
X = data[selected_features]
y = data["Loan_Status"]
```

```python
# Split numerical and categorical features
numerical_features = X.select_dtypes(include=['float64', 'int64']).columns.tolist()
categorical_features = ['Education']
data
```

```python
# One-Hot encode categorical features
encoder = OneHotEncoder(drop='first', sparse=False)  # Drop first to avoid multicollinearity
X_encoded = encoder.fit_transform(X[categorical_features])
```

```python
# Get unique feature names for the encoded categorical features
unique_values = data['Education'].unique()
encoded_names = [f'{categorical_features[0]}_{value}' for value in unique_values[1:]]
```

```python
# Create a DataFrame for the encoded categorical features
X_encoded_df = pd.DataFrame(X_encoded, columns=encoded_names)
```

```python
# Concatenate numerical and encoded categorical features
X_processed = pd.concat([X[numerical_features], X_encoded_df], axis=1)
```

```python
# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_processed, y, test_size=0.3, random_state=42)
```

```python
# Initialize the AdaBoost Classifier
adaboost = AdaBoostClassifier(n_estimators=100, random_state=42)  # You can adjust parameters
```

```python
# Train the model
adaboost.fit(X_train, y_train)
```

```python
# Predict on the test set
predictions = adaboost.predict(X_test)
```

```python
# Calculate accuracy
accuracy = accuracy_score(y_test, predictions)
print("Accuracy:", accuracy)
```

```python
# Get classification report
print("\nClassification Report:")
print(classification_report(y_test, predictions))
```
